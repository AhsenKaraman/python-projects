Colab Example

```python
#load an example dataset
from vega_datasets import data
cars = data.cars()
```

```python
cars.head()
```

```python
#plot the dataset, referencing dataframe column names
import altair as alt
alt.Chart(cars).mark_bar().encode(
x = alt.X("Miles_per_Gallon", bin = True),
y = "count()",
color = "Origin"
)
```

```python
alt.Chart(cars).mark_bar().encode(
x = "mean(Miles_per_Gallon)",
y = "Origin",
color = "Origin"
)
```

```python
alt.Chart(cars).mark_point().encode(
x = "Horsepower",
y = "Miles_per_Gallon",
color = "Origin"    
).interactive()
```

Vectors/Matrix/Tensor

```python
x = np.array(7)
x
```

```python
#it is an array that's why dim is 0
x.ndim 
```

```python
#let's define a vector
x = np.array([8, 16, 28])
x
```

```python
#since it is consist of 1 row, dim is 1
x.ndim
```

```python
#create a matrix
x = np.array([[8, 16, 28],
             [5, 15, 25],
             [9, 18, 27]])
x
```

```python
x.ndim
```

```python
#if dim is greater than 2 dim it is tensor
```

```python
#let's create tensor
x = np.array([[[7, 14, 21],
               [7, 14, 21]],
              
             [[8, 16, 24],
              [7, 16, 24]],
              
             [[10, 4, 7],
              [10, 4, 7]]])
```

```python
x.ndim
```
