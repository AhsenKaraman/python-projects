```python
#Create a new data with 'price' column only
price_data = df.filter(['price'])
```

```python
#Converting the data to array
dataset = price_data.values
```

```python
import math
```

```python
#Compute the #rows to train the model 
training_data_len = math.ceil( len(dataset) *.8) 
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
#Scale the all data between 0 and 1 
scaler = MinMaxScaler(feature_range=(0, 1)) 
scaled_data = scaler.fit_transform(dataset)
```

```python
#Create scaled training data 
train_data = scaled_data[0:training_data_len  , : ]
```

```python
#Spliting data into x_train and y_train
x_train=[]
y_train = []
for i in range(60,len(train_data)):
    x_train.append(train_data[i-60:i,0])
    y_train.append(train_data[i,0])
```

```python
#Converting x_train and y_train to array
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
#Reshape the data into the shape accepted by the LSTM
x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1],1))
```

```python
#Build the LSTM network model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True,input_shape=(x_train.shape[1],1)))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dense(units=25))
model.add(Dense(units=1))
```

```python
#Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
#Train the model
model.fit(x_train, y_train, batch_size=1, epochs=1)
```

```python
#Test data set
test_data = scaled_data[training_data_len - 60: , : ]
```

```python
#Create x_test and y_test data
x_test = []
y_test =  dataset[training_data_len : , : ] 
for i in range(60,len(test_data)):
    x_test.append(test_data[i-60:i,0])
```

```python
#Convert x_test to array 
x_test = np.array(x_test)
```

```python
#Reshape the data into shape accepted for LSTM model
x_test = np.reshape(x_test, (x_test.shape[0],x_test.shape[1],1))
```

```python
#Getting the models predicted price values
predictions = model.predict(x_test) 
predictions = scaler.inverse_transform(predictions)#Undo scaling
```

```python
type(predictions)
```

```python
#Calculate/Get the value of RMSE
rmse=np.sqrt(np.mean(((predictions- y_test)**2)))
rmse
```

```python
#Ploting the data 
train = data[:training_data_len]
validation = data[training_data_len:]
validation['Predictions'] = predictions
#Visualize the data
plt.figure(figsize=(16,8))
plt.title('Model')
plt.xlabel('Date', fontsize=18)
plt.ylabel('Price USD ($)', fontsize=18)
plt.plot(train['price'])
plt.plot(validation[['price', 'Predictions']])
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()
```
